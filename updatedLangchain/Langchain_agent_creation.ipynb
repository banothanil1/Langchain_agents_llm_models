```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ['GROQ_API_KEY'] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.agents import create_agent

def get_whether(city: str) -> str:
    """Get the whether for a city."""
    return f"The whether in city {city} is sunny .."

agent = create_agent(
    model = 'groq:llama-3.3-70b-versatile',
    tools = [get_whether],
    system_prompt = "you are a helpfull assistant."
)
response = agent.invoke({"messages": [{"role": "user", "content": "what is the temparature in Hyderabad"}]})

response['messages'][-2].content
```
